**# 5) Decidan cómo abordar los valores faltantes y cómo hacerlo. (Esto requiere sus criterios, así que expliquen su elección).**

```python
# primero muestro el df filtrando todos los valores NaN, en este caso salen ambas columnas con datos nulos y en total son 3873 filas con datos NaN. 
df[df.isna().any(axis=1)]
```

```python
# Encontrar datos para weigth y size
# Defino una función para encontrar cualquier valor nulo tanto de item weight como de oulet size, teniendo en cuenta el identificar único para cada uno
# De esta manera, el valor nulo se completará con la moda e acuerdo a los datos de su identificador.
# En este caso, de las 3873 filas con datos nulos, aplicado el apply, quedan 2414 filas con datos nulos.
def encontrar(x):
  if pd.isnull(x['Item_Weight']):
    id = x['Item_Identifier']
    weight_items = df[(df['Item_Identifier'] == id) & (df['Item_Weight'].notna())]
    if len(weight_items) > 0:
      x['Item_Weight'] = weight_items['Item_Weight'].mode()[0]
  
  if pd.isnull(x['Outlet_Size']):
    id_outlet = x['Outlet_Identifier']
    size_items = df[(df['Outlet_Identifier'] == id_outlet) & (df['Outlet_Size'].notna())]
    if len(size_items) > 0:
      x['Outlet_Size'] = size_items['Outlet_Size'].mode()[0]

  return x

df = df.apply(encontrar, axis=1)
df[df.isna().any(axis = 1)]
```

```python
# Identifico cuántos valores siguen faltando
df.isna().sum()
```

```python
# Dado que no existe asociación entre Item_Identifier e Item_Weight y son solo 4 filas, decido borrarlas, no va a alterar el dato final
# Para esto busco el index, de esta forma las elimino directamente
df['Item_Weight'].isna()
df.loc[pd.isna(df["Item_Weight"]), :].index
```

```python
# Elimino los index de acuedo a valores nulos en Item_Weight
df.drop([927, 1922, 4187, 5022], axis=0, inplace=True)
df.isna().sum()
```

```python
# Ahora debo evaluar como trabajar los 2410 datos de outlet size, dado que en la función anterior ningún dato fue corregido
```
