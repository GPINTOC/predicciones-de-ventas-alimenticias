```python
# 5) Decidan cómo abordar los valores faltantes y cómo hacerlo. (Esto requiere sus criterios, así que expliquen su elección).
# primero muestro el df con el filtro de cualquier valor NaN, en este caso salen ambas columnas, no encuentro nada que sea común para ellas, además en ningún caso hay dos NaN en una misma fila
df[df.isna().any(axis=1)]
```

```python
# hago el análisis solo con la columna Item_Weight con NaN
df[df['Item_Weight'].isna()]
```

```python
# hago el análisis solo con la columna Outlet_Size con NaN
df[df['Outlet_Size'].isna()]
```

```python
# imputar valores faltantes con promedio para la columna Item_Weight, ya que es un dato numerico
# Busco el promedio y reemplazo los nulos con el promedio
promedio_Item_Weight = df['Item_Weight'].mean()
print(promedio_Item_Weight)
df['Item_Weight'].fillna(promedio_Item_Weight, inplace = True)
df[df['Item_Weight'].isna()]
```

```python
# imputar valores faltantes con moda para la columna Outlet_Size
# Busco la moda, en este caso existe solo una moda por lo que se ocupa el index 0 y reemplazo los nulos con esta.
moda_Outlet_Size = df['Outlet_Size'].mode()
print(moda_Outlet_Size)
df['Outlet_Size'].fillna(moda_Outlet_Size[0], inplace = True)
df[df['Outlet_Size'].isna()]
```

```python
# 6) Confirmen que no hay valores faltantes después de abordarlos.
df.isna().sum()
```
